Following only works for single core simulations

```python
#snapshot for slip velocity
jobid=0
coord=np.loadtxt('xyz'+str(jobid)+'_0.dat')
x=coord[:,0]
ncell=len(x)

field="vel"

fp = open(str(field) + str(jobid) +'_0.dat','rb')
ary = np.fromfile(fp, np.float64, -1)
fp.close()
nt=int(ary.shape[0]/ncell)
vel=ary.reshape(nt,ncell)

plt.yscale('log')
plt.plot(x,vel[10,:])
plt.xlabel('x (km)')
plt.ylabel('slip rate (m/s)')
```

```python
#space-time plot for velocity
jobid=0
field="vel"

yr=365*24*3600

coord=np.loadtxt('xyz'+str(jobid)+'_0.dat')
x=coord[:,0]
ncell=len(x)

fp = open(str(field) + str(jobid) +'_0.dat','rb')
ary = np.fromfile(fp, np.float64, -1)
fp.close()

d=np.loadtxt('time'+str(jobid)+'.dat')
step=d[:,0]
print(len(step))
nt=int(ary.shape[0]/ncell)
d=ary[:nt*ncell].reshape(nt,ncell)
d=d.T
d=np.log10(d)


plt.figure(figsize=(8,3))
plt.pcolormesh(step,x,d,vmin=-12,vmax=0)
plt.xlabel('Time step')
plt.ylabel('x (km)')
#plt.xlim([0,3])
plt.colorbar(label='log10 slip rate(m/s)')
#plt.savefig('/home/i25004/.notebook/Swarm/stv2.png',dpi=300,bbox_inches='tight')
```

```python
#space-time plot for velocity
jobid=0
field="vel"

coord=np.loadtxt('xyz'+str(jobid)+'_0.dat')
x=coord[:,0]
ncell=len(x)


fp = open(str(field) + str(jobid) +'_0.dat','rb')
ary = np.fromfile(fp, np.float64, -1)
fp.close()

d=np.loadtxt('time'+str(jobid)+'.dat')
time=d[:,1]/yr
print(len(time))
nt=int(ary.shape[0]/ncell)
d=ary[:nt*ncell].reshape(nt,ncell)
d=d.T
d=np.log10(d)


plt.figure(figsize=(8,3))
plt.pcolormesh(time,x,d,vmin=-12,vmax=0)
plt.xlabel('Time (yr)')
plt.ylabel('x (km)')
#plt.xlim([0,3])
plt.colorbar(label='log10 slip rate(m/s)')
#plt.savefig('/home/i25004/.notebook/Swarm/stv2.png',dpi=300,bbox_inches='tight')
```
